```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numerical)
```

```python
knn = NearestNeighbors(n_neighbors = 5, algorithm='auto')
knn.fit(df_scaled)
```

```python
def find_similar_songs(track_index, n_neighbors=5):
    distances, indices = knn.kneighbors([df_scaled[track_index]], n_neighbors=n_neighbors+1)
    similar_songs_indices = indices.flatten()[1:] #excluding first song as it is the song itself
    return df.iloc[similar_songs_indices]    
```

```python
def get_artist_genres(artist_name):
    result = sp.search(q=f'artist:{artist_name}', type='artist')
    if result['artists']['items']:
        return result['artists']['items'][0]['genres']
    return []
```

```python
def get_track_genres(track_name, artist_name):
    result = sp.search(q=f'track:{track_name} artist:{artist_name}', type='track')
    if result['tracks']['items']:
        track_id = result['tracks']['items'][0]['id']
        track_info = sp.track(track_id)
        album_id = track_info['album']['id']
        album_info = sp.album(album_id)
        artist_genres = get_artist_genres(artist_name)
        album_genres = album_info.get('genres', [])
        return list(set(artist_genres + album_genres))  # Combine artist and album genres
    return get_artist_genres(artist_name)
```
